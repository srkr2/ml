```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.cluster import KMeans

Iris = pd.read_csv("Iris.csv")
X = Iris.iloc[:,[2,3]].values
modelKM = KMeans(n_clusters=3, init='k-means++',max_iter=100,n_init=10,random_state=42)
y = modelKM.fit_predict(X)
```

```python
cluster = pd.DataFrame(X,columns=['Petal Length','Petal Width'])
cluster['Species'] = y
centroids = modelKM.cluster_centers_
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.scatterplot(cluster,x='Petal Length',y='Petal Width',hue="Species",palette="deep")
plt.scatter(centroids[:,0],centroids[:,1],color='black')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
print("silhouette score: ", silhouette_score(X, y))
```
